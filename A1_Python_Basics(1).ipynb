### <span style="color: red"> Task 1:</span> The text file you downloaded earlier ('E1_T1.csv') contains ten words. Read the words into a list and use a "for loop" to print out each word. Also print the words as a full sentence (i.e., a single string), as well as the data type of the sentence.

### <span style="color: blue"> Solution:</span>

```python
#importing dependencies
import os
import csv

#getting current directory & stating file name
os.getcwd()
filename = 'data-1/E1_T1.csv'#changes depending on the directory relative to the cwd

#opening file
with open(filename, 'r', newline='') as f:
    #creating reader object
    csv_file = csv.reader(f)
    
    #reading in data
    word_list=next(csv_file)
    
    #printing out each word
    for word in word_list:
        print(word)

    #joining words together into a single string
    joined_words = ' '.join(word_list)
    
    #printing out both the sentence and its data type
    print(joined_words,'\n',f'Data type : {type(joined_words)}')
```

### <span style="color: red"> Task 2:</span> Create a function that finds the largest number in a list. Don not use built-in function. Show you function by using a random number generator to create the inputs.

### <span style="color: blue"> Solution:</span>

```python
#importing dependencies
from random import randint

#defining function
def max_value(number_list):
    try:
        #take out first value and assign it as max val
        max_val = number_list.pop()
        #compares all values and assigns the largest to max val
        for value in number_list:
            if value > max_val:
                max_val = value
        return max_val
    except:
        return print("We could'nt find the max value. There might be somethinf")

#creating a list of random integers
random_list = [randint(1,100) for x in range(10)]

#printing out results
print(f"{random_list} \n max value of the list is {max_value(random_list)}")
```

### <span style="color: red"> Task 6:</span> Who has the most runner-up (i.e., second place) finishes in the histroy of NHL? Create a table with the teams and their runner-up finishes. The table must contain the the following columns:

-Team

-Total number of runner-up finishes

-Years [list of the years they finished as a runner-up]

You have to use Jupyter for the practice and produce a notebook with good readability. Data stored in 'Stanely_cup_winners_by_year.xlsx'.

### <span style="color: blue"> Solution:</span>

```python
#first import dependencies
import pandas as pd
import numpy as np

#opening file
file = "data-1/Stanely_cup_winners_by_year.xlsx"
df = pd.read_excel(file)
df.head()
```

```python
#dropping unncessesary columns
df.drop(['Team'],axis=1,inplace=True)
df.head()
```

```python
#creating a data frame that creates the list of years a team has won.
df1 = df.groupby('Runner up')['Year'].agg(list).reset_index()
df1.head()
```

```python
#creating a data fram that has the total value counts of the times a team has been runner up
df2 = df['Runner up'].value_counts().rename_axis('Runner up').reset_index(name='Runner up count')
df2.head()
```

```python
#merging both using the 'Runner up' column
df_final = pd.merge(df2,df1,on='Runner up',how='right')
```

```python
#final table
df_final
```
